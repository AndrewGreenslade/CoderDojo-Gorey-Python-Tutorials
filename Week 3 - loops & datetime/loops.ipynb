```python
ListOfNames = []

for i in range(5):
    name = input("Whats the name you want to input into list ?")
    ListOfNames.append(name)

print(ListOfNames)
```

```python
ListOfNamesWhileLoop = []

while True:
    name = input("Whats the name you want to input into list ?")
    ListOfNamesWhileLoop.append(name)
    if ListOfNamesWhileLoop[len(ListOfNamesWhileLoop) - 1] == "Andrew":
        break;

print(ListOfNamesWhileLoop)
```

```python
ListOfNamesWhileLoop = []

while True:
    name = input("Whats the name you want to input into list ?")
    ListOfNamesWhileLoop.append(name)
    if ListOfNamesWhileLoop[len(ListOfNamesWhileLoop) - 1] == "stop":
        ListOfNamesWhileLoop.pop();
        break;

print(ListOfNamesWhileLoop)
```
